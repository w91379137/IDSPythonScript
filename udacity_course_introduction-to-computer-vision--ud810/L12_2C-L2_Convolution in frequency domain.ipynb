```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def visible(fft_imag):
    return 20 * np.log(np.abs(fft_imag) + 0.0001)

img = cv2.imread('./lena.jpg',0)

f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.subplot(122),plt.imshow(visible(fshift), cmap = 'gray')

plt.show()
```

```python
rows, cols = img.shape
crow, ccol = rows / 2 , cols / 2
len_remove = 13

fshift_edit = fshift.copy()
fshift_edit[crow - len_remove: crow + len_remove, ccol -len_remove: ccol + len_remove] = 0

f_ishift = np.fft.ifftshift(fshift_edit)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

plt.subplot(121),plt.imshow(img_back, cmap = 'gray')
plt.subplot(122),plt.imshow(visible(fshift_edit), cmap = 'gray')
plt.show()
```
