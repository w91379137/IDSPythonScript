```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
img = cv2.imread('../Opencv/ignore/face.JPG', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 255])

plt.subplot(1, 3, 1)
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap = 'gray')

plt.subplot(1, 3, 3)
plt.bar(range(len(hist)), hist, width = 1)

plt.show()
```

http://docs.opencv.org/trunk/da/d22/tutorial_py_canny.html

```python
edges = cv2.Canny(img_gray, 50, 150)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_equalizeHist = cv2.equalizeHist(img_gray)
plt.imshow(img_equalizeHist, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
from scipy.ndimage import filters
img_gaussian = filters.gaussian_filter(img_equalizeHist, 5)
plt.imshow(img_gaussian, cmap = 'gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
imx = filters.sobel(img_gray, 1)
imy = filters.sobel(img_gray, 0)
img_g = np.sqrt(imx.astype(np.float) ** 2 + imy.astype(np.float) ** 2).astype(np.uint8)

plt.subplot(1, 3, 1)
plt.imshow(imx, cmap = 'gray')

plt.subplot(1, 3, 2)
plt.imshow(imy, cmap = 'gray')

plt.subplot(1, 3, 3)
plt.imshow(255 - img_g, cmap = 'gray')

plt.show()
```

```python
imx = filters.gaussian_filter(img_gray, (5, 5), (0, 1))
imy = filters.gaussian_filter(img_gray, (5, 5), (1, 0))
img_g = np.sqrt(imx.astype(np.float) ** 2 + imy.astype(np.float) ** 2).astype(np.uint8)

plt.subplot(1, 3, 1)
plt.imshow(imx, cmap = 'gray')

plt.subplot(1, 3, 2)
plt.imshow(imy, cmap = 'gray')

plt.subplot(1, 3, 3)
plt.imshow(255 - img_g, cmap = 'gray')

plt.show()
```
